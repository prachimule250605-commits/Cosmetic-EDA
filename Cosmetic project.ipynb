```python
#Is there a relationship between price and rating?
import numpy as np

x = df['Price']
y = df['Rating']

m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(7,5))
plt.scatter(x, y)
plt.plot(x, m*x + b)   # trendline

plt.title("Price vs Rating with Trendline")
plt.xlabel("Price")
plt.ylabel("Rating")
plt.show()
```

## Inference:
#### This chart visualizes the Price-to-Quality relationship, showing whether consumers feel that "you get what you pay for" based on the slope of the trendline.Correlation: An upward slope suggests higher-priced items are perceived as better quality, while a flat line indicates that paying more doesn't necessarily guarantee a better experience.Outlier Detection: Points far above the line represent "hidden gems" (high rating, low price), while points far below identify "overpriced" products that fail to meet expectations.

```python
# Do discounts affect the rating of products?
import numpy as np

x = df['Discount']
y = df['Rating']

m, b = np.polyfit(x, y, 1)

plt.figure(figsize=(7,5))
plt.scatter(x, y, alpha=0.5)
plt.plot(x, m*x + b, color='red')  # trendline

plt.title("Discount vs Rating with Trendline")
plt.xlabel("Discount (%)")
plt.ylabel("Rating")
plt.show()
```

## Inference:
#### This scatter plot and trendline analyze the impact of discounting on customer perception, revealing if lower prices lead to higher satisfaction or cheaper brand perception.Value Perception: An upward trendline suggests discounts make customers "happier" with their purchase value, while a downward trendline could indicate that heavy discounts are associated with lower-quality products.Correlation Strength: The spread of dots around the red line shows how much "Discount" actually influences "Rating"; a tight cluster means a strong link, while a scattered cloud suggests other factors (like quality) matter much more.Optimal Pricing: It helps identify if there is a "sweet spot" where discounts are high enough to boost ratings without devaluing the product in the eyes of the consumer.

```python
# What is the correlation between numeric variables like price, rating, discount, and quantity?
numeric_cols = ['Price', 'Rating', 'Discount', 'Stock']
correlation_matrix = df[numeric_cols].corr()
correlation_matrix

import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm', fmt=".2f")
plt.title("Correlation Between Numeric Variables")
plt.show()
```

### Infernece:
#### This heatmap visualizes the strength and direction of relationships between your key metrics, revealing how price, ratings, and inventory levels influence each other.Key Dependencies: It identifies which variables move together—for example, a positive correlation between Price and Rating would suggest that premium products are generally meeting higher consumer expectations.Impact of Discounts: It highlights whether Discounts effectively drive higher Ratings or if they are primarily used to move slow-selling Stock (negative correlation with Stock).Variable Independence: Near-zero values point out independent factors, showing you which metrics (like Stock levels) operate regardless of others like Rating or Price.

```python
#How do price, rating, and discount relate to each other visually?
import seaborn as sns
import matplotlib.pyplot as plt

sns.pairplot(df[['Price', 'Rating', 'Discount']])
plt.suptitle("Pairwise Relationships: Price, Rating, Discount", y=1.02)
plt.show()
```

### Inference:
#### This pairplot provides a comprehensive view of the interactions between your three key metrics, combining individual distributions with their pairwise relationships.Multivariate Discovery: It allows you to see all correlations at once, helping you spot hidden patterns—such as whether high discounts are strictly reserved for expensive items or lower-rated ones.Data Distribution: The diagonal plots show the "shape" of your data, revealing if your prices are mostly low-end, if your ratings are naturally skewed high, or if your discounts are concentrated at specific percentages.Cluster Identification: The scatter plots help identify "product personas," such as high-price/high-rating luxury goods versus low-price/low-rating budget options.

```python
# How do ratings differ across category + skin type combinations?
import seaborn as sns
import matplotlib.pyplot as plt

# Pivot table for heatmap
rating_pivot = df.pivot_table(values='Rating', index='Category', columns='Skin_Type', aggfunc='mean')

plt.figure(figsize=(12,6))
sns.heatmap(rating_pivot, annot=True, cmap='YlGnBu', fmt=".2f")
plt.title("Average Rating by Category and Skin Type")
plt.xlabel("Skin Type")
plt.ylabel("Category")
plt.show()
```

### Inference:
#### This heatmap identifies the Product-Market Fit by showing which specific categories are most successful for different skin types.Top Performers: The darkest cells represent the "sweet spots"—the specific combinations of category and skin type that receive the highest customer satisfaction.Quality Gaps: Lighter cells highlight areas where products may be underperforming, suggesting a need for better formulations tailored to those specific skin types.Niche Success: It reveals if certain brands or categories are specialized (e.g., a brand that excels at "Dry Skin" products but struggles with "Oily Skin" types).
